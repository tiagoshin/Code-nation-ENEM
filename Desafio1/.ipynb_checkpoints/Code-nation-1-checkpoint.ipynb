```python
Nota_final = ((df['NU_NOTA_MT']*3 + df['NU_NOTA_CN']*2 + df['NU_NOTA_LC']*1.5 + df['NU_NOTA_CH']*1 +  df['NU_NOTA_REDACAO']*3)/10.5)
```

```python
def notafinal(notas):
    notafinal = (notas[0]*3 + notas[1]*2 + notas[2]*1.5 + notas[3]*1 + notas[4]*3)/(3+2+1.5+1+3)
    return notafinal
```

```python
df['NOTA_FINAL'] = df[['NU_NOTA_MT','NU_NOTA_CN','NU_NOTA_LC','NU_NOTA_CH','NU_NOTA_REDACAO']].apply(notafinal, axis=1)
```

```python
df['NOTA_FINAL'] = df['NOTA_FINAL'].round(1)
```
